#### Check the country codes which is not available in pycountry

```python
codes1=df1['Country'].dropna().unique()  # drop the repeated values
codes2=df2['Country'].dropna().unique()
codes1,codes2
```

```python
valid_1=set(c.alpha_3 for c in pycountry.countries )
invalid_1=[ c for c in codes1 if c not in valid_1 ]
invalid_2=[c for c in codes2 if c not in valid_1]
invalid_1, invalid_2
```

```python
df1[df1['Country'].isin(invalid_1)]['Country'].value_counts()
```

```python
special_cases = {
    'GLOBAL': 'Global',
    'WB_LI': 'Low Income',
    'WB_LMI': 'Lower Middle Income',
    'WB_UMI': 'Upper Middle Income',
    'WB_HI': 'High Income',
    'AFR': 'African Region',
    'AMR': 'Americas Region',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'European Region',
    'SEAR': 'South‑East Asia Region',
    'WPR': 'Western Pacific Region'
}
```

```python
def code_convert(code):
    if pd.isna(code):
        return None
    code=code.strip().upper()
    country=pycountry.countries.get(alpha_3=code) # holds Current code(alpha_3='IND')
    if country:
        return country.name
    return special_cases.get(code,code)     # .get(code,code) if present in special case print or else wtever the code would there, will print it
```

```python
df1
```

```python
# Apply into Function

df1['Country']=df1['Country'].apply(code_convert)
df2['Country']=df2['Country'].apply(code_convert)
```

```python
sns.barplot(data=df1, x='Region', y='Mean_Estimate')
plt.title('Region wise Obesity estimation')
plt.xticks(rotation=45)
plt.show()
```
