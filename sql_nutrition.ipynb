```python
query_9='''select f.country, f.year, f.mean_estimate as female_obesity, m.mean_estimate as male_obesity,
(f.mean_estimate - m.mean_estimate) as difference from obesity f join obesity m on 
f.country=m.country and f.year=m.year and f.age_group =m.age_group 
where f.gender='female' = m.gender='male' and (f.Mean_Estimate - m.Mean_Estimate) > 5
order by difference desc limit 10 ;'''
obe_9=pd.read_sql_query(query_9, engine)
obe_9
```

```python
# 1. Obesity vs malnutrition comparison by country(any 5 countries)
```

```python
query_1='''select o.country, avg(o.mean_estimate) as Avg_obesity, avg(m.mean_estimate)as Avg_Malnutrition
from obesity o join malnutrition m
on o.country=m.country where o.country in ('india','pakistan','japan','Bangladesh','Sri Lanka')
group by o.country;'''
om_1=pd.read_sql_query(query_1, engine)
om_1
```

```python
# 2. Gender-based disparity in both obesity and malnutrition
```

```python
query_2='''select o.gender, avg(o.mean_estimate) as Avg_obesity, avg(m.mean_estimate)as Avg_Malnutrition
from obesity o join malnutrition m
on o.gender=m.gender  where o.gender ='female'
group by o.gender;'''
om_2=pd.read_sql_query(query_2, engine)
om_2
```

```python
query_02='''select o.gender, avg(o.mean_estimate) as Avg_obesity, avg(m.mean_estimate)as Avg_Malnutrition
from obesity o join malnutrition m
on o.gender=m.gender  where o.gender ='male'
group by o.gender;'''
om_02=pd.read_sql_query(query_02, engine)
om_02
```

```python
# 3. Region-wise avg estimates side-by-side(Africa and America)
```

```python
query_3='''select o.region, avg(o.mean_estimate) as Avg_obesity, avg(m.mean_estimate)as Avg_Malnutrition
from obesity o join malnutrition m
on o.region=m.region where o.region in('africa','americas')
group by o.region;'''
om_3=pd.read_sql_query(query_3, engine)
om_3
```

```python
# 4. Countries with obesity up & malnutrition down
```

```python
query_4='''SELECT o.Country,
  ROUND(MAX(o.mean_estimate) - MIN(o.mean_estimate), 2) AS Obesity_Change,
  ROUND(MAX(m.mean_estimate) - MIN(m.mean_estimate), 2) AS Malnutrition_Change
FROM obesity o JOIN malnutrition m ON o.Country = m.Country GROUP BY Country
HAVING Obesity_Change > 0 AND Malnutrition_Change < 5 ORDER BY Obesity_Change DESC;'''
om_4=pd.read_sql_query(query_4, engine)
om_4
```

```python
# 5. Age-wise trend analysis
```

```python
query_5='''SELECT o.age_group,o.year,ROUND(AVG(o.mean_estimate), 2) AS avg_obesity,
ROUND(AVG(m.mean_estimate), 2) AS avg_malnutrition
FROM obesity o JOIN malnutrition m ON o.age_group = m.age_group AND o.year = m.year
GROUP BY o.age_group, o.year ORDER BY o.age_group, o.year;'''
om_5=pd.read_sql_query(query_5, engine)
om_5
```
